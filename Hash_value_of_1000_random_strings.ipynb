```python
import numpy as np
```

```python
string_lengths = np.random.randint(low=5,high=11,size=1000)
```

```python
string_lengths[0]
```

```python
def hash_value(s):
    l=len(s)
    p=11
    h=0
    for i in range(0,l):
        h = h + (ord('{}'.format(s[i]))*(p**(i+1)))
                 
    return h
```

```python
alpha_array = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```python
random_strings = list(str(' ') * 1000)
```

```python
for i in range(0,1000):
    for j in range(string_lengths[i]):
        random_strings[i] =  random_strings[i] + alpha_array[np.random.randint(low=0,high=52,size=1)[0]]
```

```python
hash_array=[]
```

```python
for i in random_strings:
    hash_array.append(hash_value(i))
```

```python
dictionary=dict(zip(random_strings,hash_array))
```

```python
sort_dict = sorted(dictionary.items(), key=lambda x: x[1], reverse=True)
```
