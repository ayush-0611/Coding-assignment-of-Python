```python
import numpy as np
```

```python
num = np.random.randint(low=20,high=100,size=1000)
```

```python
print(num)
```

```python
Dict={}
```

```python
(unique, counts) = np.unique(num, return_counts=True)
```

```python
frequencies = np.asarray((unique, counts))
```

```python
print(frequencies)
```

```python
length=len(frequencies[0])
```

```python
for i in range(0,length):
        Dict[frequencies[0][i]]=frequencies[1][i]
```

```python
print(Dict)
```
