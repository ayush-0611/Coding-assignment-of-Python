```python
import numpy as np
```

```python
num_list=[]
for i in range(1000):
    num = int(np.random.randint(low=89,high=123,size=1))
    num_list.append(num)
```

```python
#num_list
```

```python
even_list=[]
prime_list=[]
for i in num_list:
    if i%2==0:
        even_list.append(i)
    else:
        for divisor in range(2,i//2+1):
            if i%divisor==0:
                break
            if divisor==i//2 :
                prime_list.append(i)
        
```

```python
Dict=dict.fromkeys(['even','prime'])
```

```python
Dict['even']=even_list
```

```python
Dict['prime']=prime_list
```
